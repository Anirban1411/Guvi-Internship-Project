```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for heatmap

def csv_file_analyzer(file_path):
    """
    Analyzes a CSV file to provide insights such as total rows, columns,
    data types, missing values, and summary statistics.

    Parameters:
    file_path (str): The path to the CSV file to be analyzed.
    """
    # Data Input Handling
    try:
        # Specify semicolon as delimiter and replace -200 with NaN
        df = pd.read_csv(file_path, delimiter=';', na_values=-200)
        # Drop the last two columns as they are not relevant
        df.drop(df.columns[-2:], axis=1, inplace=True)
        print(f"Successfully read the CSV file: {file_path}")
    except Exception as e:
        print(f"Error reading the CSV file: {e}")
        return

    # Total number of rows and columns
    total_rows, total_columns = df.shape
    print(f'Total Rows: {total_rows}')
    print(f'Total Columns: {total_columns}')

    # Column names and their data types
    print("\nColumn Names and Data Types:")
    print(df.dtypes)

    # Data Cleaning & Preprocessing
    # Count of missing values in each column
    missing_values = df.isnull().sum()
    print("\nMissing Values Count:")
    print(missing_values)

    # Convert relevant columns to numeric and fill missing values with the mean
    for column in df.columns:
        if column not in ['Date', 'Time']:  # Exclude 'Date' and 'Time' columns
            df[column] = pd.to_numeric(df[column], errors='coerce')
            df[column].fillna(df[column].mean(), inplace=True)

    # Logic/Analysis Implementation
    # Summary statistics for numerical columns
    print("\nSummary Statistics for Numerical Columns:")
    summary_stats = df.select_dtypes(include=[np.number]).describe()
    print(summary_stats)

    # Visualization of missing values
    plt.figure(figsize=(10, 6))
    plt.title('Missing Values Heatmap')
    sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
    plt.show()

    # Visualization of summary statistics
    summary_stats.loc['mean'].plot(kind='bar', title='Mean of Numerical Columns')
    plt.ylabel('Mean Value')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout to prevent labels overlapping
    plt.show()

# Example usage
if __name__ == "__main__":
    file_path = '/content/drive/MyDrive/AirQualityUCI.csv'  # Replace with your CSV file path
    csv_file_analyzer(file_path)
```
